Splitting the data into training data & Testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
# importing tensorflow and Keras

import tensorflow as tf # DL library we can easily create neural networks throught it 
tf.random.set_seed(3) # random way of generating values 
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                            keras.layers.Flatten(input_shape=(30,)), # convert the data into single dimensional array
                            keras.layers.Dense(20,activation='relu'), # 20 neurons in inner layer 
                            # keras.layers.Dense(30,activation='sigmoid')
                            keras.layers.Dense(2,activation='sigmoid') # 2 neurons in output layer (no. of neurons in Output layer = no. of classes)
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
            loss = 'sparse_categorical_crossentropy',
            metrics = ['accuracy'])
```

```python
# Training the Neural Network

history = model.fit(X_train_std, Y_train , validation_split=0.1 , epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['tarining data','validation data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['tarining data','validation data'],loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
